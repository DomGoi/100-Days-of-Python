```python
#Guess the number
logo = '''

 _______  __   __  _______  _______  _______    _______  __   __  _______    __    _  __   __  __   __  _______  _______  ______
|       ||  | |  ||       ||       ||       |  |       ||  | |  ||       |  |  |  | ||  | |  ||  |_|  ||  _    ||       ||    _ |
|    ___||  | |  ||    ___||  _____||  _____|  |_     _||  |_|  ||    ___|  |   |_| ||  | |  ||       || |_|   ||    ___||   | ||
|   | __ |  |_|  ||   |___ | |_____ | |_____     |   |  |       ||   |___   |       ||  |_|  ||       ||       ||   |___ |   |_||_
|   ||  ||       ||    ___||_____  ||_____  |    |   |  |       ||    ___|  |  _    ||       ||       ||  _   | |    ___||    __  |
|   |_| ||       ||   |___  _____| | _____| |    |   |  |   _   ||   |___   | | |   ||       || ||_|| || |_|   ||   |___ |   |  | |
|_______||_______||_______||_______||_______|    |___|  |__| |__||_______|  |_|  |__||_______||_|   |_||_______||_______||___|  |_|

'''
import random
from IPython.display import clear_output


def number_random_gen():
    return random.randint(1, 101)

def easy_mode():
    lives = 10
    number_think = number_random_gen()

    game_on = True
    while game_on:
        guess_number = int(input("Guess the number: \n").strip())

        if guess_number > number_think and lives != 0:
            lives -= 1
            print(f"Try lower. \nLives left: {lives}")

        elif guess_number < number_think and lives != 0:
            lives -= 1
            print(f"Try higher. \nLives left: {lives}")

        elif guess_number != number_think and lives == 0:
            print(f"You lose. The number was {number_think}. \nLives left: {lives}")
            game_on = False

        else:
            print("You win.")
            game_on = False

def medium_mode():
    lives = 7
    number_think = number_random_gen()

    game_on = True
    while game_on:
        guess_number = int(input("Guess the number: \n").strip())

        if guess_number > number_think and lives != 0:
            lives -= 1
            print(f"Try lower. \nLives left: {lives}")

        elif guess_number < number_think and lives != 0:
            lives -= 1
            print(f"Try higher. \nLives left: {lives}")

        elif guess_number != number_think and lives == 0:
            print(f"You lose. The number was {number_think}. \nLives left: {lives}")
            game_on = False

        else:
            print("You win.")
            game_on = False

def hard_mode():
    lives = 5
    number_think = number_random_gen()

    game_on = True
    while game_on:
        guess_number = int(input("Guess the number: \n").strip())

        if guess_number > number_think and lives != 0:
            lives -= 1
            print(f"Try lower. \nLives left: {lives}")

        elif guess_number < number_think and lives != 0:
            lives -= 1
            print(f"Try higher. \nLives left: {lives}")

        elif guess_number != number_think and lives == 0:
            print(f"You lose. The number was {number_think}. \nLives left: {lives}")
            game_on = False

        else:
            print("You win.")
            game_on = False

final_loop = True
while final_loop:
    print(logo)
    print("Welcome to the Number Guessing Game !")
    print('I am thinking of a number between 1 and 100.')

    difficulty_level = input("Choose a difficulty. Type 'easy', 'medium' or 'hard': ").lower().strip()  # lives 10,7,5

    if difficulty_level == 'easy':
        easy_mode()
        final_loop = False

    elif difficulty_level == 'medium':
        medium_mode()
        final_loop = False

    elif difficulty_level == 'hard':
        hard_mode()
        final_loop = False

    else:
        print("Invalid input, please try again.")
        final_loop = False

    restart = input("Do you want to play again? Type 'y' or 'n': ").lower().strip()
    if restart == 'n':
        print("Thank you for playing.")
        final_loop = False
    elif restart == 'y':
        final_loop=True
        clear_output()
    elif restart != 'y':
        print('Invalid input.')
```
