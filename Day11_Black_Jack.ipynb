```python
import random
import os

def clear():
    os.system('cls' if os.name == 'nt' else 'clear')

logo = """
.------.            _     _            _    _            _
|A_  _ |.          | |   | |          | |  (_)          | |
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   <
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |
      `------'                           |__/
"""

print(logo)
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

def user_cards():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    list_ycard = []
    list_ycard.append(cards[random.randint(0, 12)])
    list_ycard.append(cards[random.randint(0, 12)])

    for i in list_ycard:
        if i == 11:
            print(list_ycard)
            ace = int(input("Do you want the Ace to be 11 or 1? ").strip())

            # Validate the user input for Ace
            while ace not in [1, 11]:
                print("Invalid input. Please enter 1 or 11.")
                ace = int(input("Do you want the Ace to be 11 or 1? ").strip())

            list_ycard[list_ycard.index(i)] = ace
            print("Your cards after choosing Ace:", list_ycard)

    return list_ycard

def comp_cards():
    # Computer cards random
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    list_ccard = []
    list_ccard.append(cards[random.randint(0, 12)])
    list_ccard.append(cards[random.randint(0, 12)])

    while sum(list_ccard) < 17:
        list_ccard.append(cards[random.randint(0, 12)])

    return list_ccard

# Program running
blackjack = True

while blackjack:
    play = input('Do you want to play blackjack? Type "y" or "n": ').lower().strip()

    if play == 'y':
        ycards = user_cards()
        ccards = comp_cards()
        print(f'Your cards: {ycards}')
        print(f'Dealer cards: {ccards[0]}')

        if sum(ccards) > 21:
            print(f'Dealer cards: {ccards}')
            print('You win, dealer\'s over 21.')

        draw_card = input('Do you want to get another card? Type "y" or "n": ').lower().strip()

        if draw_card == 'y':
            ycards.append(cards[random.randint(0, 12)])
            print(f'Your cards after drawing another card: {ycards}')
            print(f'Computer cards: {ccards}')
            sum_ycards = sum(ycards)
            sum_ccards = sum(ccards)

            if sum_ycards > sum_ccards:
                print(f'Your cards: {ycards}')
                print(f'Computer cards: {ccards}')
                print("You win.")
            elif sum_ycards < sum_ccards:
                print(f'Your cards: {ycards}')
                print(f'Computer cards: {ccards}')
                print('You lose.')
            elif sum_ycards == sum_ccards:
                print(f'Your cards: {ycards}')
                print(f'Computer cards: {ccards}')
                print('It is a tie.')
            else:
                print('Something went wrong.')

            print('Thank you for playing.')
            blackjack = False
            clear()

        elif draw_card == 'n':
            sum_ycards = sum(ycards)
            sum_ccards = sum(ccards)

            if sum_ycards > sum_ccards:
                print(f'Your cards: {ycards}')
                print(f'Computer cards: {ccards}')
                print("You win.")
            elif sum_ycards < sum_ccards:
                print(f'Your cards: {ycards}')
                print(f'Computer cards: {ccards}')
                print('You lose.')
            elif sum_ycards == sum_ccards:
                print(f'Your cards: {ycards}')
                print(f'Computer cards: {ccards}')
                print('It is a tie.')
            else:
                print('Something went wrong.')

            print('Thank you for playing.')
            blackjack = False
            clear()

        else:
            print('Invalid input. Please enter "y" or "n".')

    elif play == 'n':
        print("Thank you for choosing simple blackjack.")
        blackjack = False
        clear()
        break

    else:
        print('Invalid input. Please enter "y" or "n".')
```
