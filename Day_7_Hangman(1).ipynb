```python
#Hangman
logo='''
 _
| |
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |
                   |___/    '''

print(logo)
from IPython.display import clear_output
#import nltk
import random
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
# Download the words dataset from NLTK
#nltk.download('words')

# # Get a list of English words
# word_list = nltk.corpus.words.words()

# # Filter out proper nouns and words with non-alphabetic characters
# word_list = [word.lower() for word in word_list if word.isalpha()]

# # Set the seed for reproducibility
# random.seed(42)

# # Choose a subset of words for the hangman game
# chosen_words = random.sample(word_list, k=500)




def play_game():
  chosen_word = random.choice(word_list)

  #Testing code
  #print(f'Pssst, the solution is {chosen_word}.')
  lives=6
  display=[]
  word_len=len(chosen_word)

  for n in range(word_len):
    display += '_'




  while "_" in display and lives >0:
    print(f"{' '.join(display)}")
    guess = input("Guess a letter: ").lower()
    clear_output()
    if guess in display:
      print(f"You have already tried: {guess}.")
      #clear()
    correct_guess=False
    for position in range(word_len):
        letter=chosen_word[position]
        if letter == guess:
          display[position]=letter
          correct_guess=True
    if not correct_guess:
      lives -= 1
      print("That is not in the word, you lose a life.")
      print(stages[lives])

  if "_" not in display:
    print("Congratulations! You have won.")
  else:
    print("Sorry, you ran out of lives. The word was:",chosen_word)

while True:
  play_game()
  restart=input('Do you want to play again? (y/n):').lower()
  if restart != 'y':
    print("Thank you for playing!")
    break
```
